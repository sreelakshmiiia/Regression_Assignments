```python
import pandas as pd
```

```python
dataset=pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
independent= dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent= dataset[['charges']]
```

```python
independent
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.2,random_state=0)
```

```python
#applying standarisation
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
# preprocessing output
scy=StandardScaler()
y_train=scy.fit_transform(y_train)
y_test=scy.transform(y_test)
```

```python
x_train
```

```python
y_train
```

```python

from sklearn.svm import SVR 
# regressor=SVR(kernel='sigmoid',C=10000)
#using non linear poly
regressor=SVR(kernel='poly')
regressor.fit(x_train,y_train)
```

```python
# regressor.coef_
```

```python
y_pred=regressor.predict(x_test)
y_pred                        
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
r2
```

```python
import pickle
#save input standard scalar
pickle.dump(sc,open("input_scalar.pkl",'wb'))
#save the model    
pickle.dump(regressor,open("final_model.pkl",'wb'))
#Save output standard scalar
pickle.dump(scy,open("output_scalar.pkl",'wb')) 
```
