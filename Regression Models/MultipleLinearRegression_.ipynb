```python
import pandas as pd
```

```python
#data collection
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
#data preprocessing
#As we have categorical data as input we have to convert in to binary values.
#State input has nominal data so we have to use one hot encoding, so it uses input field expansion
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
independent
```

```python
dependent=dataset[['charges']]
```

```python
dependent
```

```python
#Split the dataset into train and test set 
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#Model creation
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
#prediction 
y_pred= regressor.predict(x_test)
```

```python
#Evaluation
from sklearn.metrics import r2_score
r2score=r2_score(y_test,y_pred)
```
